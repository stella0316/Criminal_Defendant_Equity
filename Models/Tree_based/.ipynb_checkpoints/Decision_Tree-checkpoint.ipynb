```python
import pandas as pd
from sklearn import tree
from sklearn.model_selection import train_test_split
from sklearn import metrics
from matplotlib import pyplot as plt
from sklearn.model_selection import GridSearchCV
import graphviz
from IPython.display import Image
```

```python
data = pd.read_csv('../../Finalized Data/final_with_doc.csv')
```

```python
data.drop(['FirstD_month_x', 'AppealD_month_x', 'File', 'Unanimous', 'days_jail_doc'],axis=1,inplace=True)
```

```python
y = data.pop('Y')
```

```python
X = data
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size = 0.25)
```

```python
tree = tree.DecisionTreeClassifier()
```

```python
tree.fit(X_train, y_train)
```

```python
forecast = tree.predict(X_test)
```

```python
plot_roc(y_test, forecast,'Decision Tree Untuned')
plt.show()
```

```python
def plot_roc(actual,forecast,name):
    
    #ROC
    fpr, tpr, thresholds = metrics.roc_curve(actual, forecast)

    #AUC
    aucc = metrics.auc(fpr, tpr)

    plt.plot(fpr, tpr, label = name + ' AUC: '+str(round(aucc, 4)))
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title('ROC Curves')
    plt.grid(True)
    plt.legend(loc = 'lower right', title='AUC')
```

### Tuning

```python
#Grid search
param_grid = [
    {'criterion' : ['gini', 'entropy'],
     'max_depth': [10,50,None],
      'min_samples_split': [50, 100, 200],}  
]
```

```python
grid_search = GridSearchCV(tree.DecisionTreeClassifier(), param_grid=param_grid, scoring= 'roc_auc')
grid_search.fit(X_train, y_train)
```

```python
grid_search.best_params_
```

```python
best_tree = tree.DecisionTreeClassifier(criterion= 'entropy', max_depth= 10, min_samples_split= 100)
```

```python
best_tree.fit(X_train, y_train)
```

```python
forecast = best_tree.predict(X_test)
```

```python
plot_roc(y_test, forecast,'Decision Tree Tuned')
plt.show()
```

```python
tree.export_graphviz(best_tree, out_file='tree_classifier.dot')
```
