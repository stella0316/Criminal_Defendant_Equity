```python
import re

def years(df):
    num_entries = df.shape[0]
    output = []
    for i in range(num_entries):
        row = df[i]
        if pd.isnull(row):
            output.append('')
            continue
        row = row.replace(',','')
        temp = re.findall('....?', row)
        out = ';'.join(temp).replace(';0000','')
        output.append(out)
    return output
```

```python
j = years(main.DA_contested_years)
```

```python
main['DA_contested_years'] = j
```

```python
main = main.merge(names,how='left',left_index=True, right_on='File')
```
