```python
import pandas as pd
```

```python
judge_index = pd.read_csv(r'csv/judge_index.csv')
```

```python
judge_info = pd.read_csv(r'csv/judges_final.csv')
```

```python
judge_index.head()
```

```python
judge_index['File'].duplicated().any()
```

```python
judge_info.head()
```

```python
judges_distinct_last_name = judge_info[judge_info['judges_first_name'].isnull()]
judges_same_last_name = judge_info[~judge_info['judges_first_name'].isnull()]
```

```python
len(judges_same_last_name) + len(judges_distinct_last_name) == len(judge_info)
```

```python
judges_same_last_name['judges_first_name'] = judges_same_last_name['judges_first_name'].apply(lambda x:x.strip())
judges_same_last_name['judges_last_name'] = judges_same_last_name['judges_last_name'].apply(lambda x:x.strip())
judges_distinct_last_name['judges_last_name'] = judges_distinct_last_name['judges_last_name'].apply(lambda x:x.strip())
```

    Step1: Merge by last name, inner join

```python
merge_1 = judge_index.merge(judges_distinct_last_name,left_on='Last_Name',right_on='judges_last_name',how = 'inner')
```

```python
judge_index2 = judge_index[~judge_index['File'].isin(merge_1['File'])]
```

    Step2: Merge by first name and last name, left join

```python
merge_2 = judge_index2.merge(judges_same_last_name,left_on=['First_Name','Last_Name'],right_on=['judges_first_name','judges_last_name'],how = 'left')
```

```python
len(merge_1) + len(merge_2) == len(judge_index)
```

```python
merged = merge_1.append(merge_2)
```

```python
merged.columns
```

```python
del_column = ['Unnamed: 0_x','Unnamed: 0_y','judges_last_name','judges_first_name','ADA']

for col in del_column:
    del merged[col]
```

```python
merged.head(10)
```

```python
len(merged)
```

```python
merged.to_csv(r'csv/merged_judges.csv')
```

```python
df = pd.read_csv(r'csv/merged_judges.csv')
```

```python
df.head(10)
```

```python
df.isnull().sum()
```

```python
df['law school'].unique()
```

```python
#if law school is missing, indicate as 1
df['law school'] = df['law school'].fillna('unknown')
df['judge_law_school_missing'] = df['law school'].apply(lambda x: 1 if x == 'unknown' else 0)
```

```python
del df['grad year']
```

```python
#if elect is missing, indicate as 1
df['judge_elect_missing'] = df['elect'].fillna('unknown').apply(lambda x: 1 if x == 'unknown' else 0)
```

```python
df['judge_military_missing'] = df['military'].fillna('unknown').apply(lambda x: 1 if x == 'unknown' else 0)
```

```python
df['judge_APD_missing'] = df['APD'].fillna('unknown').apply(lambda x: 1 if x == 'unknown' else 0)
```

```python
df['judge_info_missing'] = df['nycourts'].apply(lambda x: 1 if pd.isnull(x) else 0)
```

```python
del df['ballotpedia']
```

```python
df['judge_Prof_missing'] = df['Prof'].fillna('unknown').apply(lambda x: 1 if x == 'unknown' else 0)
```

```python
df['judge_CC_missing'] = df['CC'].apply(lambda x: 1 if pd.isnull(x) else 0)
```

```python
df['judge_SC-AJ_missing'] = df['SC-AJ'].apply(lambda x: 1 if pd.isnull(x) else 0)
df['judge_SC_missing'] = df['SC'].apply(lambda x: 1 if pd.isnull(x) else 0)
df['judge_SC-AJ & SC_missing'] = df['SC-AJ & SC'].apply(lambda x: 1 if pd.isnull(x) else 0)
```

```python
df['start year'] = df['start year'].fillna(0)
df['start year'] = df['start year'].apply(lambda x:0 if x == 'Unknown' else x)
df['judge_startyear_missing'] = df['start year'].apply(lambda x: 1 if x == 0 else 0)
df['start year'] = df['start year'].apply(lambda x: int(x))
```

```python
df['judge_female_missing'] = df['female'].fillna('unknown').apply(lambda x: 1 if x == 'unknown' else 0)
```

```python
df = df.fillna(0)
```

```python
df = df.select_dtypes(include=['float64']).astype(int)
```

```python
for col in df.columns:
    if df[col].dtype == 'float64':
        df[col] = df[col].astype(int)
```
