### Adding contested marker

```python
def election_year(column):
    '''
    Takes a column and extracts year and indicator of whether a candidate was contested, 
    and if a candidate lost
    '''
    n = len(column)
    years = np.zeros(n)
    contested = np.zeros(n)
    lost = np.zeros(n)
    for i in range(n):
        row = column[i]
        try:
            year = int(row[0:4]) #try to extract row. If it's not a year, skip
            years[i] = year
            if 'uncontested' in row:
                contested[i] = 0
            elif 'contested' in row:
                contested[i] = 1
            elif 'lost' in row:
                contested[i] = 1
                lost[i] = 1
        except (TypeError, ValueError):
            continue
    return years, contested, lost
            
```

```python
#Extract year and whether contested from data
years, contested, lost = election_year(working_df.contested)
```

```python
working_df['contested_years'] = years
```

```python
#First pass over contested years
years_contested = {}
for year in working_df.contested_years:
    cont_year = np.zeros(len(working_df.contested_years))
    if year >0:
        for i in range(len(working_df.contested_years)):
            if working_df.contested_years[i] == year:
                cont_year[i] = contested[i]
        years_contested[year] = cont_year
```

```python
#Extract year and whether contested from data - part II, extra columns
years2, contested2, lost2 = election_year(working_df['Unnamed: 11'])
```

```python
def update_contested_years(years, new_years, new_contested):
    '''
    Update years dictionary with new entries
    '''
    n = len(new_years)
    for i in range(n):
        year = new_years[i]
        if year>0:
            try:
                years[year][i] = new_contested[i]
            except KeyError:
                years[year] = np.zeros(n)
                years[year][i] = new_contested[i]
        
    return years
```

```python
update1 = update_contested_years(years_contested, years2, contested2)
```

```python
#Extract year and whether contested from data - part III, extra columns
years3, contested3, lost3 = election_year(working_df['Unnamed: 12'])
```

```python
update2 = update_contested_years(update1, years3, contested3)
```

```python
#Perform the rest of updates

for k in range(12):
    col = 'Unnamed: '+str(12+1)
    years4,contested4,lost4 = election_year(working_df[col])
    update = update_contested_years(update2,years4,contested4)
    update2 = update
```

```python
#Sort years
sorted_years = sorted(update2.keys())
```

```python
#Add contested years information to the dataframe
for year in sorted_years:
    print(year)
    working_df[str(int(year))+'-contested'] = update2[year]
```
