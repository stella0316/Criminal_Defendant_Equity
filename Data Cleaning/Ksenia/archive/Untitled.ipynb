```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('final_DA_key.csv')
```

```python
data.drop('da_county',axis=1,inplace=True)
```

```python
data.head()
```

```python
data['da_female_nan'] = np.zeros(data.shape[0])
```

```python
data['da_female_nan'][data.da_female.isnull()] = 1
```

```python
data.head()
```

```python
def add_nan(data,column_name,shape):
    data[str(column_name)+'_nan'] = np.zeros(shape)
    data[str(column_name)+'_nan'][data[column_name].isnull()] = 1
```

```python
add_nan(data, 'da_university', data.shape[0])
```

```python
data.head()
```

```python
#data['election_nan'] = np.zeros(data.shape[0])
data['election_nan'][data.da_2017.isnull()] = 1
```

```python
data['da_contested_nan'] = np.zeros(data.shape[0])
```

```python
for year in range(2004,2015):
    try:
        data['contested_nan'][data['da_'+str(year)+'-contested'].isnull()] = 1
    except KeyError:
        continue
```
