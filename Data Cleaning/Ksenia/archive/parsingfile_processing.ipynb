### 1. 

Append all the parsed files together. 

```python
import pandas as pd
import numpy as np
import os
```

```python
data_dir = 'NY-Appellate-Scraping/2017-09-24/Archive/parsing20032015/'
data_filename = os.listdir(data_dir)
```

```python
data1315 = pd.DataFrame()
for name in range(0, 11):
    filename = data_dir + data_filename[name]
    file = pd.read_table(filename, sep=',', error_bad_lines=False)
    data1315 = data1315.append(file)
```

```python
data1315.shape
```

```python
data_dir = 'NY-Appellate-Scraping/2017-09-24/Archive/parsing2017/'
data_filename = os.listdir(data_dir)
data17 = pd.DataFrame()
for name in range(len(data_filename)):
    filename = data_dir + data_filename[name]
    file = pd.read_table(filename, sep=',', error_bad_lines=False, encoding = "ISO-8859-1")
    data17 = data17.append(file)
```

```python
data_dir = 'NY-Appellate-Scraping/2017-09-24/Archive/parsing20152016/'
data_filename = os.listdir(data_dir)
data1516 = pd.DataFrame()
for name in range(len(data_filename)):
    filename = data_dir + data_filename[name]
    file = pd.read_table(filename, sep=',', error_bad_lines=False, encoding = "ISO-8859-1")
    data1516 = data1516.append(file)
```

```python
data = pd.concat([data1315, data1516, data17])
```

```python
data.columns
```

Code outcome variable. There are four types of outcomes: affirmed, modified, dismissed, reversed. Anything other than affirmed indicates there are some errors done by the lower court, which is of interest. So coding every affirmed cases as 0 and other cases as 1. 

```python
data['Keywords'].unique()
```

```python
outcome0 = ['affirmed', 'affirmed;interest of justice', 'affirmed;plea', 'affirmed;interest of justice;plea', 
           'affirmed;waiver of appeal;plea', 
           'affirmed;interest of justice;waiver of appeal;plea', 'affirmed;waiver of appeal']
outcome_q = ['interest of justice;plea', 'interest of justice', 'waiver of appeal;plea', 
             'plea', 'interest of justice;waiver of appeal;plea']
```

```python
data.shape
```

```python
data = data.drop(columns = ['index'])
```

```python
data.tail()
```

```python
data[data['Keywords'].isin(outcome_q)].shape
```

```python
index = data[data['Keywords'].isin(outcome_q)].index
```

Delete all the records with outcome_q. Code all the outcome_0 into 0 and others as 1. 

```python
data = data.drop(index = index)
data.shape
```

```python
data['Keywords'].unique()
```

Convert the date into weekdays and weekends.

```python
data.loc[36708, 'FirstDate']
```

```python
data['First_Date'] = pd.to_datetime(data['FirstDate'])
data['Appeal_Date'] = pd.to_datetime(data['AppealDate'])
```

```python
data
```

```python
data['FirstD_dow'] = data.First_Date.dt.weekday_name
data['AppealD_dow'] = data.Appeal_Date.dt.weekday_name
data['FirstD_weekday'] = [0 if x in ['Saturday','Sunday'] else 1 for x in data['FirstD_dow']]
data['AppealD_weekday'] = [0 if x in ['Saturday','Sunday'] else 1 for x in data['AppealD_dow']]
```

```python
data['FirstD_month'] = data.First_Date.dt.month
data['AppealD_month'] = data.Appeal_Date.dt.month
```

```python
data['Y'] = [0 if x in outcome0 else 1 for x in data['Keywords']]
```

```python
data.to_csv("criminal_date.csv")
```
